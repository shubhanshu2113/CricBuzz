```python
import requests
import json

API_KEY = "13c92f16f8msh3cc3f324fca2fdep11155djsn7ee675a36724"
BASE_URL = "https://cricbuzz-cricket.p.rapidapi.com"

headers = {
    "X-RapidAPI-Key": API_KEY,
    "X-RapidAPI-Host": "cricbuzz-cricket.p.rapidapi.com"
}

formats = ["test", "odi", "t20"]

all_stats = {}

for fmt in formats:
    params = {"matchType": fmt}
    resp = requests.get(f"{BASE_URL}/stats/v1/topstats/0", headers=headers, params=params)

    if resp.status_code == 200:
        data = resp.json()
        headers_list = data.get("headers", [])
        rows = data.get("values", [])[:10]

        players = []
        for row in rows:
            values = row.get("values", [])
            player_dict = dict(zip(["rank"] + headers_list, values))
            players.append(player_dict)

        all_stats[fmt] = players
        print(f"✅ Got top 10 for {fmt.upper()}")
    else:
        print(f"❌ Failed for {fmt.upper()} — {resp.status_code}")

with open("top10_run_scorers_all_formats.json", "w", encoding="utf-8") as f:
    json.dump(all_stats, f, indent=4)

print("🎉 Saved data to top10_run_scorers_all_formats.json")
```

```python
import requests
import json
import time

API_KEY = "5194cf99aamsh2d1109a24f2a064p13f389jsn19e5cd52ab52"
BASE_URL = "https://cricbuzz-cricket2.p.rapidapi.com"

headers = {
    "X-RapidAPI-Key": API_KEY,
    "X-RapidAPI-Host": "cricbuzz-cricket2.p.rapidapi.com"
}

def collect_all_series():
    categories = ["international", "domestic"]
    all_series = []

    for cat in categories:
        url = f"{BASE_URL}/series/v1/{cat}"
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            series_blocks = response.json().get("seriesMapProto", [])
            for block in series_blocks:
                month_group = block.get("date")  # e.g., "FEBRUARY 2024"
                for s in block.get("series", []):
                    all_series.append({
                        "seriesId": s.get("id"),
                        "seriesName": s.get("name"),
                        "startDate": s.get("startDt"),
                        "endDate": s.get("endDt"),
                        "monthGroup": month_group,
                        "category": cat
                    })
            print(f"✅ Collected {len(all_series)} series so far from {cat}")
        else:
            print(f"❌ Failed for {cat} → {response.status_code}")

        # 🔹 Pause to avoid hitting rate limits
        time.sleep(2)

    # Save to JSON file
    with open("all_series.json", "w", encoding="utf-8") as f:
        json.dump(all_series, f, indent=2)

    print(f"\n🎉 Done! Total series collected: {len(all_series)}")

if __name__ == "__main__":
    collect_all_series()
```

```python
import requests
import json

API_KEY = "13c92f16f8msh3cc3f324fca2fdep11155djsn7ee675a36724"
BASE_URL = "https://cricbuzz-cricket.p.rapidapi.com"

headers = {
    "X-RapidAPI-Key": API_KEY,
    "X-RapidAPI-Host": "cricbuzz-cricket.p.rapidapi.com"
}

def get_stats_types():
    url = f"{BASE_URL}/stats/v1/topstats"
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print("❌ Failed to fetch stats types:", response.status_code)
        return []

    data = response.json()
    stat_types = []
    for block in data.get("statsTypesList", []):
        for t in block.get("types", []):
            stat_types.append({
                "value": t.get("value"),
                "header": t.get("header"),
                "category": t.get("category")
            })
    return stat_types

def fetch_stats(stat_type):
    url = f"{BASE_URL}/stats/v1/topstats/0"
    params = {"statsType": stat_type}
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"❌ Failed for {stat_type}: {response.status_code}")
        return {}

if __name__ == "__main__":
    all_data = {}

    stat_types = get_stats_types()
    if not stat_types:
        print("❌ No stat types found")
        exit()

    print(f"✅ Found {len(stat_types)} stat types")

    for stype in stat_types:
        value = stype["value"]
        print(f"📊 Fetching {stype['header']} ({value})")
        data = fetch_stats(value)
        all_data[value] = data

    # Save everything
    with open("player_stats.json", "w", encoding="utf-8") as f:
        json.dump(all_data, f, indent=4)

    print(f"🎉 Player stats saved in player_stats.json with {len(all_data)} categories")

```
